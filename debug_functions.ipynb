# Отладка функций репозиториев
Здесь можно тестировать любые функции по отдельности

```python
import pymongo
from datetime import datetime, timedelta
from src.config.db_config import MONGO_URI_PROD, MONGO_URI_STAGE, DB_NAME
from src.repositories.users_repository import find_users_with_birthday

# Выберите окружение
ENVIRONMENT = 'prod'  # 'prod' или 'stage'
```

```python
# Подключение к БД
mongo_uri = MONGO_URI_PROD if ENVIRONMENT == 'prod' else MONGO_URI_STAGE
client = pymongo.MongoClient(mongo_uri)
db = client[DB_NAME]
print(f"Connected to {ENVIRONMENT.upper()}")
```

```python
# Тест: Найти пользователей с днем рождения сегодня
today = datetime.now()
users = find_users_with_birthday(db, today)

print(f"Найдено: {len(users)}")
if users:
    print("\nПервые 5:")
    for user in users[:5]:
        print(f"  - {user.get('fullName', 'N/A')} (ID: {user['_id']})")
```

```python
# Закрыть соединение
client.close()
print("Connection closed")
```
